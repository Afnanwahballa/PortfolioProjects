```python
# Import libararies

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib. pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configurations of the plots we will create 

# Read in the data 

df = pd.read_csv(r'C:\Users\afnan\Downloads\movies.csv')
```

```python
# Lets look at the data
df.head()
```

```python
# is there any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df = df.dropna()
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
# Create correct year column
df['yearcorrect'] = df['released'].str.extract(pat='([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
#choose which column to sort by in this case 'gross'

df = df.sort_values(by= ['gross'], inplace=False, ascending=False)
```

```python
# if i want to look at all of the data
pd.set_option('display.max_rows', None)
```

```python
#drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.corr(method='pearson') 
# will only work on numerical data
# there are different types of correlations there is 'pearson', 'kendall' and 'spearman'
#and they will all give you slightly different results. kendall probably gives more than slightly a differnt result
# they all have there different way of determining correlation
#be aware of which one i am using by default and resrach why i am using them.
#pearsons usually runs by default.
#Pearson when lookina at correlation in the table budget and gross look like they have a positive correlation that is high 0.740247
#votes vs gross with a correlation 0.614751
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for numeric feature')

plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name]= df_numerized[col_name].cat.codes
            
df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for numeric feature')

plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.show()
```

```python
#the company has a pretty low correlation
# but again you can see that the correlation between budget and gross in high at 0.74, second is votes anf gross
```

```python
df_numerized.corr()
```

```python
#unstacking

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
